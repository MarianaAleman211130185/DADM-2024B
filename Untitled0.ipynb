```python
import matplotlib as plt
import numpy as np #Trabajar datos umericos de Phyton
from scipy import misc#Libreria de estadistica
import matplotlib.pyplot as plt#Hacer gr√°ficos
import urllib.request #Solicitar imagenes de internet
from PIL import Image #Presenatr imagenes
import cv2 as cv2 #Analisis de imagenes
import matplotlib.image as img #Presentar imagenes
import pandas as pd #Datasets
import seaborn as sb
```

```python
I = Image.open("palitos.jpg")
```

```python
plt.imshow(np.asarray(I))
plt.show()
```

```python
grises= I.convert('L')
```

```python
plt.imshow(np.asarray(grises))
plt.show()
```

```python
n, m = np.asarray(grises).shape
print(n)
print(m)
```

```python
#Umbralizado de forms manual
th=80
g = np.zeros_like(np.asarray(grises),dtype=int)
for x in range (n):
  for y in range (m):
    i = np.asarray(grises)[x, y]
    if i >= th:
      g[x, y] = 1
    else:
      g[x, y] = 0
#Carga a la nueva matriz g todos los valores que se compraron con el umbrelado
#La matriz g solamente tiene valor entre 0 y 1
```

```python
plt.imshow(g, cmap='gray')
plt.show()
```

```python
#Impeimir el histoframa de la imagen esca√±la de grises
plt.hist(np.asarray(grises).ravel(),256,[0,256]);
plt.show()
```

```python
from scipy.ndimage import binary_opening, binary_closing
```

```python
se1 = np.array([[0, 1, 0],
                [1, 1, 1],
                [0, 1, 1]], dtype=bool)
```

```python
#La imagen binaria es "g" se le aplica un elemento estructural
ge1 = binary_opening(g,structure=se1)
```

```python
plt.figure()
plt.imshow(ge1, cmap='gray')
plt.show()
```

```python
# Create a disk-shaped structuring element with radius 4
radius = 3
se2 = np. zeros ((2 * radius + 1, 2 * radius + 1), dtype=bool)
y, x = np.ogrid[-radius: radius + 1, - radius: radius + 1]
mask = x**2 + y**2 <= radius**2
se2[mask] = 1
#Perform binary opening on 'g' using the structuring
ge3 = binary_closing(ge1, structure=se2)
```

```python
plt.figure()
plt.imshow(ge3, cmap='gray')
plt.show()
```
